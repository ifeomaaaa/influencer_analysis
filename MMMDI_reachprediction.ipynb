```python
import pandas as pd
```

```python
df = pd.read_csv("Melswardrobe.csv", encoding="utf-8")
```

```python
df.head()
```

```python
df.drop(["caption", "imgUrl", "videoUrl", "id", "profileUrl", "query", "error", "message", "cookieError", "timestamp"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.pubDate= pd.to_datetime(df.pubDate).dt.date
```

```python
#renaming description to caption

df.rename(columns = {'description':'caption'}, inplace = True)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['pubDate']) #transformed to datetime again?
```

```python
df.head()
```

```python
#add day of the week column
df["dotw"] =  df['date'].dt.day_name()
```

```python
df.head()
```

```python
df['caption'] = df['caption'].astype(str) #convert to str to run code below
```

```python
import re
def cleaner(txt):
  txt = re.sub(r'@[A-Za-z0-9_]+', '',txt)
  txt = re.sub(r'https?:\/\/[A-Aa-a0-9\.\/]+', '', txt)
  return txt

df["clean"] = df["caption"].map(lambda x: cleaner(x))
```

```python
df.head()
```

```python
df.drop(["pubDate"], axis=1, inplace=True)
```

```python
df["hashtags"] = df.caption.apply(lambda x: [x for x in x.split(" ") if x.startswith("#")])
```

```python
df.sort_values(["date"], 
                    axis=0,
                    ascending=[True], 
                    inplace=True)

df.head()
```

```python
#add engagement column likes + comments 
df['engagement'] = df["likeCount"] + df["commentCount"]
df.head()
```

```python
df['avg_engagement'] = df.groupby('dotw').engagement.transform('mean')
df.head()
```

```python
plt.bar(df["dotw"], df["avg_engagement"], color="green")
plt.title("Average engagement per day of the week") 
plt.xlabel("day of the week")
plt.ylabel("average engagement") 
plt.xticks(rotation = 90)
plt.show()
```

```python
#add impressions column engagement * her followers
df['estimated_impressions'] =  df["engagement"] * 290000
df.head()
```

```python
#predicting reach following https://thecleverprogrammer.com/2022/03/22/instagram-reach-analysis-using-python/

from sklearn.model_selection import train_test_split

x = np.array(df[['likeCount', 'commentCount', 'avg_engagement']])
y = np.array(df["estimated_impressions"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
                                                test_size=0.2, 
                                                random_state=42)
```

```python
from sklearn.linear_model import PassiveAggressiveRegressor

model = PassiveAggressiveRegressor()
model.fit(xtrain, ytrain)
model.score(xtest, ytest)
```
